```python
import yfinance as yf
import pandas as pd

stocks = {
    "AAPL": "USA",       
    "MSFT": "USA",       
    "BMW.DE": "EU",     
    "AIR.PA": "EU",     
    "BABA": "Asian",    
    "TSM": "Asian"     
}

start_date = "2020-01-01"
end_date = "2023-12-31"

tock_dataframes = {}

for ticker, market in stocks.items():
    print(f"Downloading data for {ticker}...")
    try:
        data = yf.download(ticker, start=start_date, end=end_date, interval="1d")
        data['Ticker'] = ticker
        data['Market'] = market
        data.index = pd.to_datetime(data.index).tz_localize(None)
        stock_dataframes[ticker] = data
    except Exception as e:
        print(f"Error downloading data for {ticker}: {e}")

for ticker, df in stock_dataframes.items():
    file_name = f"{ticker}_data.csv"
    df.reset_index(inplace=True)  
    df.rename(columns={"index": "Date"}, inplace=True)
    df.to_csv(file_name, index=False)
    print(f"Data for {ticker} saved to {file_name}")
```

```python
import threading
import queue
import time
import uuid
from functools import lru_cache
import pymysql


# Set connection parameters for the MySQL database

db_config = {
    'host': 'localhost',
    'user': 'root',
    'password': 'azh1129aomg',
    'database': 'stock_project',
    'cursorclass': pymysql.cursors.DictCursor
}


# Cache mechanism

@lru_cache(maxsize=100)
def query_with_cache(query):
    conn = pymysql.connect(**db_config)
    with conn.cursor() as cursor:
        cursor.execute(query)
        result = cursor.fetchall()
    conn.close()
    return result

def query_without_cache(query):
    conn = pymysql.connect(**db_config)
    with conn.cursor() as cursor:
        cursor.execute(query)
        result = cursor.fetchall()
    conn.close()
    return result


# Asynchronous buffer
buffer_queue = queue.Queue()  # Thread-safe buffer queue
stop_event = threading.Event()  # stop background threads


def async_writer(batch_size=10, interval=1):
    # Background thread: Periodically writes buffer data to the database
    conn = pymysql.connect(**db_config)
    while not stop_event.is_set():  # Check whether the thread needs to be stopped
        try:
            batch = []
            while len(batch) < batch_size and not buffer_queue.empty():
                batch.append(buffer_queue.get())

            if batch:
                with conn.cursor() as cursor:
                    for data in batch:
                        cursor.execute(data)
                conn.commit()
                print(f"Async batch written: {len(batch)} records")
        except pymysql.MySQLError as e:
            print(f"Error during async write: {e}")
        time.sleep(interval)  # Check buffer per second
    conn.close()

# Start background thread
writer_thread = threading.Thread(target=async_writer, daemon=True)
writer_thread.start()

def write_to_async_buffer_with_deduplication(data):
    # Write data to the buffer queue and add deduplication logic.
    if data not in list(buffer_queue.queue):  # Check whether the same data already exists in the buffer
        buffer_queue.put(data)
    else:
        print(f"Skipped duplicate data: {data}")


def stop_async_writer():
    #  Stop asynchronous write threads.
    stop_event.set()
    writer_thread.join()


# target node data synchronization logic (combined with asynchronous buffers)

def replicate_data_with_buffer(source_table, target_table, delay=0, iterations=5):
    for _ in range(iterations):
        time.sleep(delay)  # network delay
        query = f"""
        INSERT INTO {target_table}
        SELECT * FROM {source_table}
        WHERE updated_at >= NOW() - INTERVAL 1 MINUTE
        ON DUPLICATE KEY UPDATE
            adj_close = VALUES(adj_close),
            close = VALUES(close),
            high = VALUES(high),
            low = VALUES(low),
            open = VALUES(open),
            volume = VALUES(volume);
        """
        write_to_async_buffer_with_deduplication(query)
    print(f"Synchronized to buffer: {source_table} -> {target_table}")


def clear_test_data():
    conn = pymysql.connect(**db_config)
    try:
        with conn.cursor() as cursor:
            cursor.execute("DELETE FROM stock_data_us;")
            cursor.execute("DELETE FROM stock_data_us_target;")
            cursor.execute("DELETE FROM stock_data_eu;")
            cursor.execute("DELETE FROM stock_data_eu_target;")
            cursor.execute("DELETE FROM stock_data_asian;")
            cursor.execute("DELETE FROM stock_data_asian_target;")
        conn.commit()
        print("All test data cleared successfully.")
    except pymysql.MySQLError as e:
        print(f"Error clearing test data: {e}")
    finally:
        conn.close()

def generate_unique_ticker(base="AAPL"):
    unique_suffix = uuid.uuid4().hex[:4]  
    return f"{base[:5]}_{unique_suffix}" 


# TEST Function

def test_cache_performance_repeated():
    query = "SELECT * FROM stock_data_us_target WHERE ticker = 'AAPL'"
    # without cache
    start_time = time.time()
    for _ in range(100):
        query_without_cache(query)
    end_time = time.time()
    print(f"100 queries without cache took {end_time - start_time:.2f} seconds.")
    # with cache
    start_time = time.time()
    for _ in range(100):
        query_with_cache(query)
    end_time = time.time()
    print(f"100 queries with cache took {end_time - start_time:.2f} seconds.")


def test_high_concurrency():
    clear_test_data()  
    threads = []
    start_time = time.time()
    for i in range(50):  # Simulate 50 threads to submit tasks concurrently
        ticker = generate_unique_ticker(base=f"AAPL{i}")
        thread_data = f"""
        INSERT INTO stock_data_us_target (date, adj_close, close, high, low, open, volume, ticker, market)
        VALUES (CURDATE(), {150+i}, {155+i}, {160+i}, {145+i}, {150+i}, 1000000, '{ticker}', 'USA')
        """
        t = threading.Thread(target=write_to_async_buffer_with_deduplication, args=(thread_data,))
        threads.append(t)
        t.start()
    for t in threads:
        t.join()
    end_time = time.time()
    print(f"High concurrency test completed in {end_time - start_time:.2f} seconds.")


def test_dynamic_buffer():
    clear_test_data() 
    buffer_sizes = [10, 100]  # buffer size are 10 and 100
    for size in buffer_sizes:
        global writer_thread, stop_event
        stop_event.set()
        writer_thread.join()
        stop_event.clear()
        writer_thread = threading.Thread(target=async_writer, args=(size,), daemon=True)
        writer_thread.start()

        start_time = time.time()
        for i in range(100):  
            ticker = generate_unique_ticker(base=f"AAPL{i}")
            sql = f"""
            INSERT INTO stock_data_us_target (date, adj_close, close, high, low, open, volume, ticker, market)
            VALUES (CURDATE(), {150+i}, {155+i}, {160+i}, {145+i}, {150+i}, 1000000, '{ticker}', 'USA')
            """
            write_to_async_buffer_with_deduplication(sql)
        time.sleep(2)  
        end_time = time.time()
        print(f"Buffer size {size}: Took {end_time - start_time:.2f} seconds.")


if __name__ == "__main__":
    print("Testing cache performance with repeated queries:")
    test_cache_performance_repeated()

    print("\nTesting high concurrency:")
    test_high_concurrency()

    print("\nTesting dynamic buffer sizes:")
    test_dynamic_buffer()

    stop_async_writer()  
```
